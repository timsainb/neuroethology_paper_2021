```python
from pathlib2 import Path
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import re
import IPython
from tqdm.autonotebook import tqdm
```

```python
A = 10
K = -0.2
C = 3
x = np.linspace(0.01, 10)
relatedness = A * np.exp(K * x) + C
```

```python
fig, ax = plt.subplots(figsize=(6,6))
ax.plot(x, relatedness, lw=4, color="k")
ax.axhline(C, lw="3", ls="dashed", color="k", alpha=0.5)

x_pos = 0
y_pos = 0
ax.arrow(
    x[-1],
    relatedness[-1],
    dx=x[-1] - x[-2],
    dy=(relatedness[-1] - relatedness[-2])/1.5,
    width=0.1,
    head_width=0.35,
    head_length=0.35,
    color="k",
    #head_starts_at_zero=True
)
[i.set_linewidth(4) for i in ax.spines.values()];

ax.text(x = 5, y=C - 0.5, s='chance', fontstyle='italic', fontsize=14, verticalalignment='center')

ax.set_ylim([0, A + C])
ax.set_xlim([x[0], x[-1] + 1])
ax.set_xlabel("Sequential Distance", fontsize=18)
ax.set_ylabel("Relatedness", fontsize=18)
```
